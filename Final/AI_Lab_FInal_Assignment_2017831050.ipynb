```python
# Importing Modules
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# Processing File and Dropping Columns
df = pd.read_csv('bank-data/bank-full.csv', delimiter=';')
to_drop = ['contact', 'day', 'month']
df = df.drop(columns=to_drop)
Y = df['y']
Y = Y
X = df.drop('y', axis=1)
```

```python
X
```

```python
# marital Column Cleanup
# single = 0
# married = 1
# divorced = 2
for index, item in enumerate(X['marital']):
  if str(item) == 'single':
    X['marital'][index] = 0
  elif str(item) == 'married':
    X['marital'][index] = 1
  elif str(item) == 'divorced':
    X['marital'][index] = 2

# poutcome Column Cleanup
# unknown = 0
# success = 1
# other = 2
# failure = 3
for index, item in enumerate(X['poutcome']):
  if str(item) == 'unknown':
    X['poutcome'][index] = 0
  elif str(item) == 'success':
    X['poutcome'][index] = 1
  elif str(item) == 'other':
    X['poutcome'][index] = 2
  elif str(item) == 'failure':
    X['poutcome'][index] = 3

# education Column Cleanup
# unknown = 0
# primary = 1
# secondary = 2
# tertiary = 3
for index, item in enumerate(X['education']):
  if str(item) == 'unknown':
    X['education'][index] = 0
  elif str(item) == 'primary':
    X['education'][index] = 1
  elif str(item) == 'secondary':
    X['education'][index] = 2
  elif str(item) == 'tertiary':
    X['education'][index] = 3

# job Column Cleanup
# unknown = 0
# primary = 1
# secondary = 2
# tertiary = 3
for index, item in enumerate(X['job']):
  if str(item) == 'unknown':
    X['job'][index] = 0
  elif str(item) == 'admin.':
    X['job'][index] = 1
  elif str(item) == 'retired':
    X['job'][index] = 2
  elif str(item) == 'unemployed':
    X['job'][index] = 3
  elif str(item) == 'management':
    X['job'][index] = 4
  elif str(item) == 'blue-collar':
    X['job'][index] = 5
  elif str(item) == 'technician':
    X['job'][index] = 6
  elif str(item) == 'services':
    X['job'][index] = 7
  elif str(item) == 'student':
    X['job'][index] = 8
  elif str(item) == 'housemaid':
    X['job'][index] = 9
  elif str(item) == 'entrepreneur':
    X['job'][index] = 10
  elif str(item) == 'self-employed':
    X['job'][index] = 11

# default Column Cleanup
# no = -1
# yes = 1
for index, item in enumerate(X['default']):
  if str(item) == 'no':
    X['default'][index] = -1
  else:
    X['default'][index] = 1

# housing Column Cleanup
# no = -1
# yes = 1
for index, item in enumerate(X['housing']):
  if str(item) == 'no':
    X['housing'][index] = -1
  else:
    X['housing'][index] = 1

# loan Column Cleanup
# no = -1
# yes = 1
for index, item in enumerate(X['loan']):
  if str(item) == 'no':
    X['loan'][index] = -1
  else:
    X['loan'][index] = 1

# Y Vector Cleanup
# no = -1
# yes = 1
for index, item in enumerate(Y):
  if str(item) == 'no':
    Y[index] = -1
  else:
    Y[index] = 1
Y = Y.astype('int')
```

```python
# Normalization
X = (X -np.min(X))/(np.max(X)-np.min(X)).values
X
```

```python
# Correlation HeatMap
corr = df.corr()
f, ax = plt.subplots(figsize=(12, 10))
mask = np.triu(np.ones_like(corr, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, annot=True, mask = mask, cmap=cmap)
```

```python
# Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
# Using 'sigmoid' kernel SVM
classifier = svm.SVC(kernel='sigmoid')
classifier = classifier.fit(x_train, y_train)
```

```python
# Confusion Matrix
matrix = ConfusionMatrixDisplay.from_estimator(classifier, x_test, y_test, cmap=plt.cm.Blues, normalize='true')
plt.title('Confusion matrix for our classifier')
plt.show(matrix)
plt.show()
```

```python
# Accuracy on Test Data
y_pred = classifier.predict(x_test)
accuracy = accuracy_score(y_test, y_pred) * 100.0
print("Accuracy for the SVM is : ",accuracy)
```
